AdventureWorks2017

Get the top 20 unique names of products that exists in both Production.Product and Procution.WorkOrder and its total aggregation count

```sql
USE AdventureWorks2017;

 WITH reverseCte AS(
    SELECT Distinct top 20 PP.Name as reverseName, COUNT(*) as CountryAggregationCount
    FROM Production.WorkOrder as PW
        Inner join Production.Product as PP
        ON pw.ProductID = pp.ProductID
        WHERE PW.OrderQty > 3
        GROUP BY PP.Name
  
      )

SELECT reverseName, CountryAggregationCount
FROM reverseCte


```

Get the top 10 unique names of products in reverse that exists in both Production.Product and Procution.WorkOrder and its total aggregation count

```sql
USE AdventureWorks2017;

WITH reverseCte AS(
    SELECT Distinct top 10 PP.Name as reverseName, COUNT(*) as CountryAggregationCount
    FROM Production.WorkOrder as PW
        Inner join Production.Product as PP
        ON pw.ProductID = pp.ProductID
        WHERE PW.OrderQty > 3
        GROUP BY PP.Name
  
      )

SELECT REVERSE(reverseName), CountryAggregationCount
FROM reverseCte



```

AdventureWorksDW2017

Get the rows of column that has EnglishProductName, and the highest values of DealerPrice, and the SalesAmount from the Dbo.DimProducts table and join all of Dbo.DimProducts and the commonality with Dbo.FactResellerSales and group it by EnglishProductName and SalesAmount

```sql
USE AdventureWorksDW2017;

WITH getProductSalesCTE AS(
    SELECT dp.EnglishProductName AS EPN, dp.DealerPrice AS DealerPrice, frs.SalesAmount AS SalesAmount
    FROM Dbo.DimProduct as dp
    LEFT OUTER JOIN Dbo.FactResellerSales as frs
    ON dp.ProductKey = frs.ProductKey
)

SELECT EPN, MAX(DealerPrice) as HighestDealerAggregate, SalesAmount
FROM getProductSalesCTE
GROUP BY EPN, SalesAmount
ORDER BY HighestDealerAggregate DESC
```

Get the rows of column that has EnglishProductName, and the total sum values of DealerPrice, and the SalesAmount from the Dbo.DimProducts table and join all of Dbo.DimProducts and the commonality with Dbo.FactResellerSales and group it by EnglishProductName and SalesAmount that exceeds 10,000 dollars

```sql
USE AdventureWorksDW2017;

WITH getProductSalesCTE AS(
    SELECT Dp.EnglishProductName AS EPN, Dp.DealerPrice AS DealerPrice, frs.SalesAmount AS SalesAmount
    FROM Dbo.DimProduct AS Dp
    LEFT OUTER JOIN Dbo.FactResellerSales AS Frs
    ON Dp.ProductKey = frs.ProductKey
)

SELECT EPN, SUM(DealerPrice) AS HighestDealerAggregate, SalesAmount
FROM getProductSalesCTE
GROUP BY EPN, SalesAmount
HAVING SalesAmount > 10000
ORDER BY HighestDealerAggregate DESC
```

### Northwinds2022TSQLV7

Display all the order year in which ALL the current unique customers and the COMMONALITY OF THE previous years unqiue customers and then group it by the order year. Finally display the difference in growth between current years unique customers and the previous years customers

```sql
USE Northwinds2022TSQLV7;

WITH YearlyCountCTE AS
(
  SELECT YEAR(orderdate) AS orderyear,
    COUNT(DISTINCT CustomerId) AS numcusts
  FROM Sales.[Order]
  GROUP BY YEAR(orderdate)
)
SELECT Cur.orderyear, 
  Cur.numcusts AS UniqueCurrentCustomers, Prv.numcusts AS UniquePreviousCustomers,
  Cur.numcusts - Prv.numcusts AS CustomerGrowth
FROM YearlyCountCTE AS Cur
  LEFT OUTER JOIN YearlyCountCTE AS Prv
    ON Cur.orderyear = Prv.orderyear + 1;
```

Display only the order year in which the current unique customers and the COMMONALITY OF previous years unqiue customers and then group it by the order year. Finally display the difference in growth between current years unique customers and the previous years customers

```sql
USE Northwinds2022TSQLV7;

WITH YearlyCountCTE AS
(
  SELECT YEAR(orderdate) AS orderyear,
    COUNT(DISTINCT CustomerId) AS numcusts
  FROM Sales.[Order]
  GROUP BY YEAR(orderdate)
)
SELECT Cur.orderyear, 
  Cur.numcusts AS UniqueCurrentCustomers, Prv.numcusts AS UniquePreviousCustomers,
  Cur.numcusts - Prv.numcusts AS CustomerGrowth
FROM YearlyCountCTE AS Cur
  INNER JOIN YearlyCountCTE AS Prv
    ON Cur.orderyear = Prv.orderyear + 1;
```

WorldWideImporters

Display TOP 100 rows of an unique count of customer id where there is commonality between the Sales.Customer Table and the Sales.Orders Table, and display the AmountExcludingTax, and the all unique TaxAMount that is not 0 also group it by the Exclusion of tax and the current tax

```sql
USE WideWorldImporters;

WITH FindCustomerNamesCTE AS (
SELECT TOP 100 COUNT(DISTINCT Cust.CustomerID) AS UniqueCustomerCount, Cust.AmountExcludingTax AS exTax, TaxAmount AS Taxed
FROM Sales.CustomerTransactions as CusT
GROUP BY AmountExcludingTax, TaxAmount
HAVING TaxAmount != 0.0

)

SELECT DISTINCT TOP 100 exTax, Taxed, UniqueCustomerCount
FROM FindCustomerNamesCTE AS Cus
INNER JOIN Sales.Orders AS Sales
ON CustomerID = Sales.CustomerID 

```

### PrestigeCarsOriginal

Display the customerName and the unique count of it with the TotalSalePrice and SaleDate only within commonality between the Sales.Customer table and the data.Sales table and finallu group the aggregation count by customer name , total saleprice and the sale date

```sql
USE PrestigeCarsOriginal;

WITH CustomerNameCountCTE AS (
Select cus.CustomerName as cn, Count(cus.CustomerName) as UniqueNameCount, s.TotalSalePrice as TotalSalePrice, s.SaleDate as SaleDate
from data.Customer as cus 
INNER JOIN data.Sales as s
on cus.CustomerID = s.CustomerID
GROUP BY Cus.CustomerName, S.TotalSalePrice, S.SaleDate

)


SELECT cn as CustomerName, UniqueNameCount, TotalSalePrice, SaleDate
FROM CustomerNameCountCTE



```
